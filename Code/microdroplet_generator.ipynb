```python
# SINGLE CHANNEL DATA
R_single= 8* 0.9 *(10**6) *(1/0.1 + 1/0.6)**4  /np.pi
print(R_single/10**10)

# no. of channels= n_c
n_c = [20+i for i in range(40)]
```

```python
mu = 10**-3
constant_mul = mu * 10**9 / np.pi

# ------------------------------------------------------------------- #

w1=[(i+1)/10 for i in range(15)] # i
w2 = w1 # j
w3= [0.3, 0.4] #k
 
# n=[0,1,2,3,4] #number of turns Maybe use this as it is.

# ------------------------------------------------------------------- #

x_d = [1.5,2, 2.5] # a
h1 = [0.1, 0.2, 0.3] #b
w_R = [2.5,3,3.5] # c
l3 = [(i+1)/10 for i in range(11)] # d

```

```python
# ------------------------------------------------------------------- #
gamma=0
L_R=0
t=0
# Initialize the Resistance array with appropriate dimensions
Resistance = np.zeros((len(w1)))

# Populate the Resistance array
for i in range(len(w1)):
    for j in range(len(w2)):
        for k in range(len(w3)):
            for n in range(1):
                for p in range(len(n_c)):
                    for a in range(len(x_d)):
                        for b in range(len(h1)):
                            for c in range(len(w_R)):
                                for d in range(len(l3)):
                                    L_R = n_c[p]*(0.6+w3[k]) - w3[k] +1
                                    l4 = [0.5]
                                    l5=[0.5]
                                    # beta = 0.5
                                    # while  (beta + 1) < L_R:
                                    #     beta += 0.5
                                    #     l4.append(beta)
                                    #     l5.append(beta)
                                
                                    R_total=R_single/n_c[p]/10**9 

                                    for delta in range(len(l4)):    
                                        for theta in range(len(l5)):
                                           
                                            l1= l3[d] + 1.5 +1.5 +3*w2[j] +w_R[c]
                                            if (w2[j]>w1[i]):
                                                break

                                            if (n==0):
                                                if delta>0 or theta>0:
                                                    break

                                            if (n==1):
                                                if theta>0:
                                                    break

                                            if (n%2==0):
                                                gamma = n* (l4[delta]+l5[theta]+0.2)
                                            else:
                                                gamma= (n+1)*(l4[delta]+0.2)/2+ (n-1)*(l5[theta]+0.2)/2
                                            
                                            t=(l3[d]+l1+ (L_R)/2 + x_d[a] +n*(0.2*w1[i])) +gamma
                                            
                                            alpha = (8 * (t) * ((1 / w1[i] + 1 / h1[b]) ** 4))
                                            
                                            R_Y= (7 * (0.5 + 3 * (L_R -1) / 16) * ((1 / w2[j] + 1 / h1[b]) ** 4))

                                            Resistance[i] = constant_mul * (alpha + R_Y)

                                            zeta = Resistance[i]/10**12
                                            if abs(zeta-R_total)<(0.005*R_total):
                                                if ((L_R+x_d[a]+0.9)<40): #constraint that the length must be less than 40mm
                                                    if(l1 +w1[i]<6.2): #constraint that (half the width) must be less than 6.25mm
                                                        print("Resistance=", zeta)
                                                        print("R_total",R_total)
                                                        print("w1=", w1[i])
                                                        print("w2=", w2[j])
                                                        print("w3=",w3[k])
                                                        print("n_c=",n_c[p])
                                                        print("x_d=",x_d[a])
                                                        print("h1=", h1[b])
                                                        print("w_R=",w_R[c])
                                                        print("l3=",l3[d])
                                                        print("l4=",l4[delta])
                                                        print("l5=",l5[theta])
                                                        print("horizontal total width=", (L_R+x_d[a]+0.9))
                                                        print("x_R=",L_R)
                                                        print("total length=", 2*l1+ 2*w1[i])
                                                        print("l1=",2*l1)
                                                        print("")

# ------------------------------------------------------------------- #
```
